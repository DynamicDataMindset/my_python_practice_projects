```python
#Resturant bill calculator, rounded off to two decimal places

print("Welcome to my Resturant bill calculator program! Enter bill items below.\n")

starter = float(input("Starter price = "))
main_course = float(input("Main_course price = "))
dessert = float(input("Dessert price = "))
drinks = float(input("Drinks price = "))

total = float(starter+main_course+dessert+drinks)
tax = total* 0.2

rounded_total = round(total, 2)
rounded_tax = round(tax, 2)

print(f"\nTOTAL BILL: R{rounded_total}")
print(f"\n4TAX: R{rounded_tax}")
print(f"\nTOTAL PRICE TO BE PAID: R{rounded_tax+rounded_total}")




```
