```python
#first Function

def sum(num1,num2):
  result = num1 +num2
  return result

#second function

def mul(num1,num2):
  result = num1 * num2
  return result

#main program

print("Welcome to my program!!")

x = float(input("Enter your First number: "))
y = float(input("Enter your second number: "))

command = input("Please Enter SUM for Addition and MUL for multiplication: ")

if command == "SUM":
  print("Result: ",sum(x,y))
elif command == "MUL":
  print("Result: ",mul(x,y))
else:
  print("invalid Command")
```
