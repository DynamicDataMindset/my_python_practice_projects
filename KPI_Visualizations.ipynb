```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the IT KPI data
data = {
    'KPI': ['Ticket Volume', 'Resolution Time (Hours)', 'First Contact Resolution Rate (%)',
            'User Satisfaction Score (1-10)', 'System Uptime (%)', 'Security Incident Count',
            'Project Completion Rate (%)', 'IT Budget Utilization (%)'],
    'Target': [150, 4, 75, 8, 99.9, 0, 90, 25],
    'Example Value': [142, 3.8, 78, 8.2, 99.95, 1, 100, 22]
}

df = pd.DataFrame(data)

# 1. Help Desk Performance Dashboard
help_desk_kpis = ['Ticket Volume', 'Resolution Time (Hours)',
                  'First Contact Resolution Rate (%)', 'User Satisfaction Score (1-10)']
help_desk_df = df[df['KPI'].isin(help_desk_kpis)]

# Bar chart for Help Desk Performance
colors = ['green' if value >= target else 'red'
          for value, target in zip(help_desk_df['Example Value'], help_desk_df['Target'])]

plt.figure(figsize=(10, 6))
plt.bar(help_desk_df['KPI'], help_desk_df['Example Value'], color=colors)
plt.axhline(0, color='gray', linewidth=0.5)
plt.title('Help Desk Performance Dashboard')
plt.ylabel('Values')
plt.xlabel('KPI')
plt.xticks(rotation=45)
plt.savefig('help_desk_dashboard.png')
plt.show()

# 2. System Uptime Trend
plt.figure(figsize=(8, 5))
plt.plot(['January'], [99.95], marker='o', label='System Uptime (%)')
plt.axhline(y=99.9, color='r', linestyle='--', label='Target (99.9%)')
plt.title('System Uptime Trend')
plt.ylabel('System Uptime (%)')
plt.legend()
plt.savefig('system_uptime_trend.png')
plt.show()

# 3. Security Incident Analysis (Pie Chart)
resolved = 1
unresolved = 0
plt.figure(figsize=(6, 6))
plt.pie([resolved, unresolved], labels=['Resolved', 'Unresolved'], autopct='%1.1f%%', startangle=90)
plt.title('Security Incident Analysis')
plt.savefig('security_incident_analysis.png')
plt.show()

# 4. Project Management Performance (Gauge Charts)
from matplotlib.patches import Wedge

# Function for gauge chart
def gauge_chart(value, target, title, filename):
    fig, ax = plt.subplots(figsize=(6, 3))
    ax.set_aspect('equal')
    wedge = Wedge((0.5, 0.5), 0.4, 0, 180, color='green' if value >= target else 'red')
    ax.add_patch(wedge)
    ax.text(0.5, 0.1, f'{value}%', ha='center', fontsize=12)
    ax.text(0.5, 0.9, title, ha='center', fontsize=15)
    plt.axis('off')
    plt.savefig(filename)
    plt.show()

# Project Completion Rate Gauge
gauge_chart(100, 90, 'Project Completion Rate', 'project_completion_rate.png')

# IT Budget Utilization Gauge
gauge_chart(22, 25, 'IT Budget Utilization', 'it_budget_utilization.png')
```
