```python
print("Welcome to my program, Enter password!")

attempts = 1

password = input("Enter password here: ")

while password != "letmein" and attempts < 3:
  print("Incorrect Password!")
  password = input("please Try Again!")
  attempts += 1

if password == "letmein":
  print("Access Granted!")
else:
  print("Access Denied! You are out of attempts!!")

```
