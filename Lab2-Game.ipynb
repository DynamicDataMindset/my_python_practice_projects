```python
import random

def my_function():
  print("AMAZING")

#generate a random integer between 1 and 100
target_number = random.randint(1, 100)

#Rules of the game
print("Welcome to guess my number game! Guess a number between 1 and 100.")
print("You have 6 attempts to guess the number.")

attempts = 1
guess = -1
while guess != target_number and attempts <= 6:
  print(f"Attempt {attempts}:")
  attempts += 1

#Ask the user to guess a number
  guess = int(input("Enter your guess: "))

  if guess < target_number:
    print("Your guess is too low! Try again.")
  elif guess > target_number:
    print("Your guess is too high! Try again.")
  elif guess == target_number:
    print(f"Congratulations! You guessed the number {target_number} in {attempts} attempts.")
    break

my_function()

if guess != target_number:
  print(f"Sorry, you've reached the maximum number of attempts. The number was {target_number}.")



```
